### 1) Matrix Multiplication 
* r[i][j] = summation of (a[i][n]*b[n][j])) here n = no of elements in row of A and columns of B, where 
* a = matrix a
* b = matrix b
* a[n][m], b[m][o]
* r = resultant matrix of order n*o

```python
# Example of matrix multiplication 
a = [[1,2,3],[4,5,6],[7,8,9]]
b = [[3,4,5],[5,8,9],[9,9,0]]
c = []
# we will check if all of row and columns are equal :
row_a = len(a)
col_b = len(b[0])

# NOTE :  assumes that every nested array are of same order as the first one,
# this code will not work if you change the order of the nested arrays 
if row_a == col_b:
    for i in range(len(a)):
        arr = [];
        for j in range(len(b[0])):
            sum = 0
            for k in range(len(a)):
                sum += a[i][k] * b[k][j]
            arr.append(sum)
        c.append(arr)
            
else:
    print("Invalid matrix multiplication operation")

c

# NOTE : we will be using numpy or scipy for mathematical computation while building actual projects where efficiency is of most value.
```

# Linear Mapping (Linear Transformation)
* T(x)=Ax, where x is a vector and A is matrix and T(x) is function of that outputs the transformed matrix 
* Below we have given the example of linear mapping.
* origin remains constant (No change in origin)
* This transformation is represented in neural network like **transformed Vector =  [ Matrix * Vector ]**

# Affine Mapping (Linear transformation + translation)
* T(x) = Ax+b where b is translation vector
* origin is changed unless b is 0 vector
* Real world example : **Neural network, y=W(transpose) x + b then y is passed to activation function, where W is a weigt matrix x is input vector and b is the bias vector**
* Input layer = No of neurons = no of dimensions in the vector
* Weight is applied in the hidden layer of the input not in the very first layer i.e input layer.

### Difference between translation and transformation : 
* Transformation is basically any kind of change which included rotation, translation, skewing, scaling, shering
* Translation is special kind of transformation where object is shifted without resizing or any other affect 

```python
# Applying above transformation to the vector using matrix multiplication
# representing above example : 
transformationMatrix = [[1,3,5],[3,4,6],[5,6,8]]

# representing column vector  
vector = [[4],[5],[6]]

# vector is also 3*1 matrix which will be transformed

# transformation  = [x*i y*j k*z] 
transformedVector = []
for i in range(len(vector)):
    arr = []
    for j in range(len(vector[0])):
        sum = 0
        for k in range(len(vector)):
            sum += transformationMatrix[k][i] * vector[k][j]
        arr.append(sum)
    transformedVector.append(arr)


transformedVector
        
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

v = np.array([transformedVector[0],transformedVector[1],transformedVector[2]]) 

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red')


ax.set_xlabel('X (î)')
ax.set_ylabel('Y (ĵ)')
ax.set_zlabel('Z (k̂)')
ax.set_title('Transformed Vector plot')

ax.set_xlim([0, 50])
ax.set_ylim([0, 70])
ax.set_zlim([0, 100])
plt.show()

# here the origin is same 
```

### 2) Dot product in linear algebra 
* Dot product is also called scalar product where we multiply each elements of two different vectors and we add them.
* Dot product squeezes n-D vectors into scalar.
* It says the similarity of the direction between two vectors
* dotproduct = 0 vectors are perpendicular to each other
* dotproduct > 0 Angle between them is less than 90
* dotrpduct < 0 vectors are in opposite direction


```python
# Dot product example 
vec1 = [1,2,4]
vec2 = [4,5,6]
dotproduct = 0;
for i in range(len(vec1)):
    dotproduct+= vec1[i]*vec2[i]

# the angle between two vectors is less than 90
dotpro1 = 0 
vec3 = [2,4,6]
vec4 = [4,8,12]
for i in range(len(vec1)):
    dotpro1+= vec3[i]*vec4[i]
    
dotproduct,dotpro1
```

### 3) Determinant 
* Is a scalar quanity that says how the area or volumne of a matrix is changed after transformation is applied.
* det == 0 then transformation squishes everything into lower dimension, this case is also called as singular matrix 
* det < 0 transformation flips the space
* det > 0 transformation doesnot changes the orientation
* det > 1 or det < –1	It stretches the space (area/volume increases)
* Between –1 and 1	It shrinks the space
* Determinant can only be found on a square matrix

### 4) Inverse

* Id determinant is non zero it is non singular, invertible , regular

### 6) Transpose
* We have a matrix A of order m*n
* we can find a transpose of A so new matrix B will be formed
* B = A Transpose
* We interchange the rows into columns and vice versa


```python
# finding transpose of matrix
mat = [[1,2,3],[4,5,6],[7,8,9]]
m = len(mat)
n = len(mat[0])

# i is a column index and j is a row index 
tranmat = []
for i in range(n):
    arr = []
    for j in range(m):
        arr.append(mat[j][i])
    tranmat.append(arr)     
tranmat
# below is the transpose of the above matrix 
```
