# Starting with Stats :
* **Types**
### A) **Descriptive** 
1) **Measure of central tendency**   
        * **Mean**    
            mean = ∑ elements / N
        * **Weighted Mean**
        * **Trimmed Mean** 
        * **Mode**    
            mode = Element with higher frequency    
        * **Median**    
            median for odd (sorted) = Element at mid position where mid = n/2   
            median for even (sorted) = Element at mid and mid+1 / 2
2) **Measure of disperssion**      
        * **Range** : (Minumum - maximum)    
        * **Variance** :  Average of squared distance between mean and every data point, ∑(x̄-datapoint)^2/N.      
        * **Mean Absolute Deviation** : ∑|xi- x̄|/N, same as variance just abosolute in term of square.   
        * **Standard Deviation σ** : Square root of the variance i.e. root under variance , its unit is same as that of data.       
        * **Cofficient of variance** : (σ / mean) * 100%

```python
#exmaple of mean mode median 
arreven = [1,2,3,4,5,6,7,7,8,9,10]
arrodd = [6,8,10,14.15]
evensum = 0
oddsum = 0


for i in range(len(arreven)):
    if i < len(arrodd):
        oddsum+=arrodd[i]

    evensum+=arreven[i]

evenmean = evensum/len(arreven)
oddmean = oddsum/len(arrodd)
evenmean,oddmean

# similarly we can find mode and median of the given data.
```

* Corelation : cov(x,y)/ σx* σy, where σ = standard deviation, cov = co variance
