# Determinant : 
* Determinant of a matrix is a single number which tells us about different properties of a matrix.
* It tell us about Area in 2D space, Volume in 3D.
* It tell us about the linear dependency of vectors.
* Tells us if the matrix is invertible
* Tells about the rank of the matrix

### Use of determinant in machine learning : 
* To find inverse of matrix used in OLS.
* To Analyse Eigen values and Eigen vectors.
* Understand the relation between columns and rows of a matrix

### Finding determinant of a matrix : 
* For triangular matrix (both upper and lower) use Sarrus Rule i.e product of dialognal element Aii
* For other use general method 
* Can only be found on square matrix
* if det(A) = 0 it is singular else it is non-singular 

### Rank of matrix : 
* Tells about the linear dependency between different vectors.
* Hence can be used in dimentionality reduction.
* Rank only exists for square or sub square matrix 

### Trace of matrix : 
* Sum of diagonal elements .
* Trace only exists for square matrix
  

# Eigen values and Eigen Vectors : 
* Eigen value is a scalar
* Eigen vector is a vector
* Vector which doesnot go under transformation of a matrix is an Eigen vector.
* But it can be scaled so by the factor with which it is scaled is called Eigen Value.
* Eigen Values and Vectors only exists for square matrix
* Av = λv
* λ is scalar, A is transformation matrix , v is a vector which undergoes transformation.
* They are written in descending order first eigen vector will be corresponding to largest eigen value, second to second largest and so on.
* **Eigenspectrum**: Set of all eigen values of a matrix.
* **Eigenspace** : Eigenspace is the set of all vectors (along with the zero vector) that correspond to a particular eigenvalue of a matrix.

### Use of Eigen values and vectors in Machine learning : 
* Directly used in PCA , Feature extraction, Dimentionality reduction.
* Eigen values and vectors are used by google page ranking algorithms.
