# Bias Variance Trade-off 
* High variance leads to overfitting
* High bias leads to underfitting
* **We as a machine learning engineer have to take care of balancing both bias and variance such that models gives otimumum result**

# Model Optimization Techniques

# End to End Machine Learning Pipeline : 
* Identify Problem.
* Gather data.
* Prprocess and convert data into disered format.
* Feed data to Machine Learning Algorithm (When algorithm is trained over data it becomes a machine learning model).
* Test the Model against the testing dataset.
* In case of unsatisfactory results tweak some hyperparameters
* Retrain the model.
* Deploy the model.
* **Machine learning model in nothing but a collection of weight's and biases learned from training data along with mathematical function**

# Deploying machine learning model with django or any other python based framework 
* Simply export the trained model.(model gets saved in form of file with extension like .h5 depending on the framework used for traning the model)
* Load the exported file in any of the framework using pickle or some other libraries
* Feed the input in the same format as the input format used while training the model else will produce error.
* Predict the output and server the output as a httpresponse() or jsonresponse()

### We can use MLops for continious training and deployment in case of data drift OR when large amount of data is produced on regular interval.
