# Linear Algebra
* Linear algebra deals with transformation, so does that means ?
* Let's understand it geomatrically :
* We have a vector given by [1,3,4]
* Scalar : Any singular value like 1,2,3,4,5,etc. They are called scaler because when any singular value multipled by any vector scales that vector.
* Vector : Programatically they are 1D array. eg - [1,2,3]
  * Row Vector : [1,2,3]
  * Column Vector : 
  *                 [1]      
                    [2]   
                    [3]
  * Unit vector is a vector whose magnitude is 1 unit, these are called unit vectors  ^i ^j ^k.Â 
  * Basis vector are vectors which are linearly independent.
  * Linearly independent : Vectors having different directions
  * Linearly Dependent : Vectors having same direction. Saying that vectors are the just scaled but their directions are same

### Matrix Multiplication 
* r[i][j] = summation of (a[i][n]*b[n][j])) here n = no of elements in row of A and columns of B, where 
* a = matrix a
* b = matrix b
* a[n][m], b[m][o]
* r = resultant matrix of order n*o

```python
# Example of matrix multiplication 
a = [[1,2,3],[4,5,6],[7,8,9]]
b = [[3,4,5],[5,8,9],[9,9,0]]
c = []
# we will check if all of row and columns are equal :
row_a = len(a)
col_b = len(b[0])

# NOTE :  assumes that every nested array are of same order as the first one,
# this code will not work if you change the order of the nested arrays 
if row_a == col_b:
    for i in range(len(a)):
        arr = [];
        for j in range(len(b[0])):
            sum = 0
            for k in range(len(a)):
                sum += a[i][k] * b[k][j]
            arr.append(sum)
        c.append(arr)
            
else:
    print("Invalid matrix multiplication operation")

c

# NOTE : we will be using numpy or scipy for mathematical computation while building actual projects where efficiency is of most value.
```

# Why scalars are called scalar : 
* Becuase they scale and vector.
  

```python
# Scalar example 
scalar = 4
vector = [4,5,6]

scaled_vector = scalar * vector
scaled_vector = [i*scalar for i in vector]

vector,scaled_vector
# left most value is initial vector and right most is a scaled vector 
```

# Let's interpret the vector 
* v = 2i + 3j - 2k
* Geomatrical interpretation of the above vector :
* 2 units in i i.e x axis
* 3 units in j i.e  y axis
* -2 units in k i.e z axis


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

v = np.array([2, 3, -2])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red')

ax.set_xlim([0, 2])
ax.set_ylim([0, 3])
ax.set_zlim([-2, 1])


ax.set_xlabel('X (iÌ)')
ax.set_ylabel('Y (Äµ)')
ax.set_zlabel('Z (kÌ)')
ax.set_title('Vector v = 2i + 3j - 2k')

plt.show()
```

# Matrix : 
* Nothing but a transformation applied to a vector
* example of transformation given as matrix:   
     [1 ,3, 5]   
[3, 4, 9]    
[5, 6, 8]
* When we do matrix multiplication we actually combine two transformation
* vector = 4i + 5j + 6k
* When we apply transformation to a vector using matrix is transforms the vector but origin remains the same.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

v = np.array([4, 5, 6])  # 4i + 5j + 6k

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red')

# Label axes
ax.set_xlabel('X (Ã®)')
ax.set_ylabel('Y (Äµ)')
ax.set_zlabel('Z (kÌ)')
ax.set_title('Vector v = 4i + 5j + 6k')


ax.set_xlim([0, 7])
ax.set_ylim([0, 7])
ax.set_zlim([0, 7])

plt.show()
```

```python
# Applying above transformation to the vector 
# representing above example : 
transformationMatrix = [[1,3,5],[3,4,6],[5,6,8]]

# representing column vector  
vector = [[4],[5],[6]]

# vector is also 3*1 matrix which will be transformed

# transformation  = [x*i y*j k*z] 
transformedVector = []
for i in range(len(vector)):
    arr = []
    for j in range(len(vector[0])):
        sum = 0
        for k in range(len(vector)):
            sum += transformationMatrix[k][i] * vector[k][j]
        arr.append(sum)
    transformedVector.append(arr)


transformedVector
        
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

v = np.array([transformedVector[0],transformedVector[1],transformedVector[2]]) 

fig = plt.figure()
ax = fig.add_subplot(projection='3d')


ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red')


ax.set_xlabel('X (Ã®)')
ax.set_ylabel('Y (Äµ)')
ax.set_zlabel('Z (kÌ)')
ax.set_title('Transformed Vector plot')

ax.set_xlim([0, 50])
ax.set_ylim([0, 70])
ax.set_zlim([0, 100])
plt.show()
```
