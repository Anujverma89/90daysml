```python
# Starting with Machine learning models : 
# Superivised machine learning algorithm : 
# Linear regression 
```

```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegressionScratch:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.lr = learning_rate
        self.epochs = epochs
        self.m = 0  # slope
        self.c = 0  # intercept
    
    def fit(self, X, y):
        n = len(X)
        for _ in range(self.epochs):
            y_pred = self.m * X + self.c
            error = y_pred - y

            # Gradient Descent update
            dm = (2/n) * np.sum(error * X)
            dc = (2/n) * np.sum(error)
            self.m -= self.lr * dm
            self.c -= self.lr * dc

    def predict(self, X):
        return self.m * X + self.c

    def mean_squared_error(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

# Sample data
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)

# Model
model = LinearRegressionScratch(learning_rate=0.01, epochs=1000)
model.fit(X, y)
y_pred = model.predict(X)

print(f"Slope (m): {model.m}")
print(f"Intercept (c): {model.c}")
print(f"MSE: {model.mean_squared_error(y, y_pred)}")

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Prediction Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression from Scratch (Gradient Descent)')
plt.legend()
plt.grid(True)
plt.show()
```
