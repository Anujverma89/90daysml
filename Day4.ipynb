# Representing vectors geomatrically

# Intution : 
* every data point in machine learning is ultimately represented as a vector — whether it's numeric, categorical, text, image, audio, etc.
* Even if it's not originally in vector form, we convert it into a vector so that models (especially mathematical ones like linear regression, neural networks, etc.) can understand and work with it.
* Numeric data:
  * Already in vector form (e.g., [3.2, 5.1, 7.9])
* Categorical data: Converted using one-hot encoding or embeddings
* Color : 
  * Red → [1, 0, 0]
* Text data: Transformed into word/sentence embeddings
  * "I love cats" → [0.23, -0.11, 0.87, ...]
* Images: Pixels or feature maps become long vectors
  * 28x28 image → 784-dim vector
* **So since these are vectors, we can visualize them geomatrically**
* **We can find angles, distance between vectors, length of vectors**
* **This gives birth to different algorithms like clustring, SVM, Classification, PCA**

# Norms : 
* **Norms are any function on a vector space that assings length to a vactor i.e. how far the vector is from the origin**    
* Simply saying function that finds the distance of vector from origin
* Types of norms (distances) :
  * L1 Norm (Manhattan Norm) finds manhattan distance,
  * L2 Norm (Euclidean Norm) finds Euclidean distance,

# How can we relate a Analytics geometry to machine leraning : 
* We can compute the distance between two vectors, which is the basis for clustering algorithms like K-Means, where data points are assigned to the nearest cluster centroid.
* It is used in machine learning algorithms that rely on distance metrics, such as:
  * K-Nearest Neighbors (KNN)
  * Support Vector Machines (SVM)
  * Anomaly detection models
* Angles between vectors (via cosine similarity) are used in:
  * Text similarity, recommendation systems, and NLP
  * Helps determine how similar two data points (e.g., documents) are.
* Vector projections and orthogonality help in:
  * Feature extraction, dimensionality reduction (e.g., PCA)
* Linear transformations (like rotation, scaling, translation) help understand:
  * How data changes in neural networks and preprocessing
* Norms and vector magnitudes are used for:
  * Normalization and regularization techniques like L1/L2 regularization
